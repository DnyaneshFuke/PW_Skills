```python
''' 2.Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
from it and implement the area() method in both.  '''
from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    @abstractmethod
    def areaa(self):
        pass

# Derived class Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def areaa(self):
        return math.pi * self.radius ** 2

# Derived class Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def areaa(self):
        return self.length * self.width

# Create instances
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Print areas
print(f"Area of Circle: {circle.areaa():.2f}")
print(f"Area of Rectangle: {rectangle.areaa()}")

```

```python
''' 9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the
fraction as "numerator/denominator". '''
import math
class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero")
        gcd = math.gcd(numerator, denominator)  # Find GCD to simplify
        self.numerator = numerator // gcd
        self.denominator = denominator // gcd

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

# Test the class
fraction = Fraction(5, 2)
print(fraction)  
fraction2 = Fraction(4, 2)
print(fraction2)  
```

```python
''' 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors. '''
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if not isinstance(other, Vector):
            return NotImplemented  
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

# Test the class
v1 = Vector(3, 4)
v2 = Vector(1, 2)
v3 = v1 + v2  

print(v3)  
```

```python
''' 12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute
the average of the grades. '''
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades  # List of grades

    def average_grade(self):
        if not self.grades:
            return 0  # Return 0 if no grades available to avoid division by zero
        return sum(self.grades) / len(self.grades)

    def __str__(self):
        return f"Student: {self.name}, Average Grade: {self.average_grade():.2f}"

# Test the class
student1 = Student("Sanvi", [85, 90, 78, 92])
student2 = Student("Mallika", [70, 88, 94])

print(student1) 
print(student2) 
```

```python
''' 14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
and hourly rate. Create a derived class Manager that adds a bonus to the salary '''
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus
    
    def __str__(self):
        return f"Manager: {self.name}, Salary: {self.calculate_salary():.2f} (Includes Bonus)"

# Creating Employee and Manager objects
Emp1 = Employee("Sham", 500, 350)
print(Emp1.calculate_salary())  

Manager1 = Manager("Sham", 500, 350, 1000)
print(Manager1)  

```

```python
''' 18. Create a class House with attributes address and price. Create a derived class Mansion that adds an
attribute number_of_rooms.    '''
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    def get_house_info(self):
        return f"Address: {self.address}\nPrice: {self.price}"

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def get_mansion_info(self):
        house_info = self.get_house_info()  # Get the basic house info
        return f"{house_info}\nNumber of Rooms: {self.number_of_rooms}"

# Test the classes
house1 = House("123 Elm St.", 200000)
mansion1 = Mansion("456 Oak Ave.", 5000000, 20)

# Display information
print(house1.get_house_info())
print("---")
print(mansion1.get_mansion_info())

                
```
