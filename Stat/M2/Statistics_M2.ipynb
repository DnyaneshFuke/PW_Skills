#Statistics Advance Part 1

1. What’s a random variable in probability?
   - It’s a function that assigns numbers to outcomes in a random experiment.
   - It can be discrete (finite/countable values) or continuous (infinite values in a range).
   - Random variables help in modeling real-world scenarios mathematically.
   - The probability distribution describes how probabilities are assigned to different values of the variable.
   - Expected value and variance help summarize the random variable’s behavior.

2. What are the types of random variables?
   - **Discrete**: Takes specific values (like the number of heads in coin flips).
   - **Continuous**: Can take any value within a range (like height or weight).
   - Discrete random variables have probability mass functions (PMF).
   - Continuous random variables have probability density functions (PDF).
   - Examples: Number of customer arrivals (discrete), temperature readings (continuous).

3. Difference between discrete and continuous distributions?
   - **Discrete**: Probabilities assigned to specific values (e.g., Binomial, Poisson).
   - **Continuous**: Uses a density function over intervals (e.g., Normal, Uniform).
   - Discrete distributions sum to 1, while continuous distributions integrate to 1.
   - Discrete probabilities are calculated as exact values, while continuous probabilities use areas under a curve.
   - Real-world example: Rolling a die (discrete), measuring rainfall (continuous).

4. What’s a probability distribution function (PDF)?
   - It gives the probability of a random variable taking a certain value (discrete) or probability density (continuous).
   - For discrete variables, it’s called a probability mass function (PMF).
   - For continuous variables, the probability of a single point is zero; instead, probability is found over an interval.
   - PDFs are used in many fields like finance, physics, and machine learning.
   - The shape of the PDF determines the behavior of the variable.

5. How’s a CDF different from a PDF?
   - **PDF**: Probability of individual values.
   - **CDF**: Cumulative probability up to a given value.
   - The CDF always increases from 0 to 1 as it sums up probabilities.
   - For discrete variables, the CDF is a step function, while for continuous variables, it’s smooth.
   - CDFs help in calculating percentiles and probabilities over ranges.

6. What’s a discrete uniform distribution?
   - Every outcome in a finite set has equal probability (like rolling a fair die).
   - The probability of each outcome is \( \frac{1}{n} \) for \( n \) total outcomes.
   - The mean is \( \frac{a + b}{2} \) where \( a \) and \( b \) are the smallest and largest values.
   - Variance is given by \( \frac{(b-a+1)^2 - 1}{12} \).
   - Used in simulations, games, and simple random sampling.

7. Key properties of a Bernoulli distribution?
   - Only two outcomes: success (1) or failure (0).
   - Defined by a probability \( p \).
   - Mean = \( p \), Variance = \( p(1-p) \).
   - The sum of multiple Bernoulli trials follows a Binomial distribution.
   - Commonly used in binary classification and success/failure modeling.

8. What’s the binomial distribution used for?
   - Models how many successes occur in \( n \) independent trials with probability \( p \) per trial.
   - The probability mass function is given by \( P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} \).
   - Mean = \( np \), Variance = \( np(1-p) \).
   - Used in quality control, genetics, and survey analysis.
   - The binomial distribution approximates the normal distribution when \( n \) is large.

9. What’s the Poisson distribution and where’s it used?
   - Models the number of events occurring in a fixed interval at an average rate \( \lambda \).
   - Probability mass function: \( P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!} \).
   - Mean and variance are both equal to \( \lambda \).
   - Used for modeling rare events like call arrivals, traffic accidents, or mutation counts.
   - Approximates the Binomial distribution when \( n \) is large and \( p \) is small.

10. What’s a continuous uniform distribution?
    - All outcomes in a range \( [a, b] \) are equally likely.
    - The probability density function (PDF) is \( f(x) = \frac{1}{b-a} \) for \( a \leq x \leq b \).
    - Mean = \( \frac{a+b}{2} \), Variance = \( \frac{(b-a)^2}{12} \).
    - Used in random number generation and Monte Carlo simulations.
    - The CDF is a linear function increasing from 0 to 1.

11. What are the characteristics of a normal distribution?
    - Symmetrical, bell-shaped curve.
    - Mean, median, and mode are equal.
    - Defined by mean \( \mu \) and standard deviation \( \sigma \).
    - 68-95-99.7 rule for standard deviations.
    - Many natural phenomena follow this distribution.

12. What is the standard normal distribution, and why is it important?
    - Normal distribution with mean 0 and standard deviation 1.
    - Allows easy probability calculations using Z-scores.
    - Used in hypothesis testing and confidence intervals.
    - Any normal distribution can be converted to standard normal.
    - Facilitates statistical comparisons.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
    - States that sample means approach a normal distribution as sample size increases.
    - Applies regardless of population distribution.
    - Basis for many inferential statistics techniques.
    - Ensures accuracy in estimating population parameters.
    - Used in hypothesis testing and confidence intervals.

14. How does the Central Limit Theorem relate to the normal distribution?
    - Justifies normal approximation for sample means.
    - Explains why normality assumptions hold in many tests.
    - Reduces skewness effects in large samples.
    - Key reason normal distribution is widely used.
    - Helps in practical statistical modeling.
    
19. What is the relationship between a Z-score and a confidence interval?
    - Z-score determines margin of error.
    - Larger Z-scores create wider confidence intervals.
    - Confidence intervals rely on Z-values for standard normal distributions.
    - Used to assess statistical significance.
    - Helps in understanding sampling variability.

20. How are Z-scores used to compare different distributions?
    - Standardize different datasets.

21. What are the assumptions for applying the Central Limit Theorem?
    - Random sampling.
    - Independent observations.
    - Sample size sufficiently large (usually \( n \geq 30 \)).
    - Finite variance.
    - Larger samples improve approximation accuracy.

22. What is the concept of expected value in a probability distribution?
    - Weighted average of all possible values.
    - Represents long-term average outcome.
    - Calculated as \( E(X) = \sum x P(x) \) for discrete variables.
    - Helps in decision-making and predictions.
    - Used in economics, insurance, and games.

```python
# 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

def plot_cdf_discrete_uniform(a, b):
    x = np.arange(a, b+1)  # Values from a to b
    cdf_values = randint.cdf(x, a, b+1)  # Compute CDF

    plt.step(x, cdf_values, where='mid', label=f"Discrete Uniform CDF [{a}, {b}]")
    plt.xlabel('x')
    plt.ylabel('CDF')
    plt.title('Cumulative Distribution Function (CDF)')
    plt.xticks(x)
    plt.yticks(np.linspace(0, 1, b-a+1))
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()

# Example Usage
a, b = 1, 6  # Example: a fair 6-sided die
plot_cdf_discrete_uniform(a, b)
```

```python
# 14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
def z_score(x, mean, std):
    return (x - mean) / std

data_point = 1.5
mean, std = 0, 1
print(f"Z-score of {data_point}: {z_score(data_point, mean, std)}")
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset.
sample_data = np.random.normal(loc=50, scale=10, size=30)
sample_mean = np.mean(sample_data)
pop_mean = 50
pop_std = 10
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(sample_data)))
print(f"Z-statistic: {z_stat}")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result.
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
conf_int = stats.norm.interval(0.95, loc=sample_mean, scale=sample_std/np.sqrt(len(sample_data)))
print(f"95% Confidence Interval: {conf_int}")
```

```python
# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
data = np.random.normal(100, 15, 100)
mean, std = np.mean(data), np.std(data, ddof=1)
conf_int = stats.norm.interval(0.95, loc=mean, scale=std/np.sqrt(len(data)))
print(f"95% Confidence Interval for Mean: {conf_int}")
```

```python
# 22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
data = np.random.normal(50, 15, 1000)
range_val = np.ptp(data)
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1
print(f"Range: {range_val}, IQR: {iqr}")
```

```python
# 24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
def calc_skew_kurt(data):
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)
    return skewness, kurtosis

skew, kurt = calc_skew_kurt(data)
print(f"Skewness: {skew}, Kurtosis: {kurt}")
```
