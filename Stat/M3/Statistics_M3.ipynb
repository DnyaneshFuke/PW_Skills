#Statistics Advance Part 2

1. What is hypothesis testing in statistics?
   - A method to make decisions about a population based on sample data.
   - Involves testing an assumption (hypothesis) using statistical methods.
   - Helps determine if an observed effect is real or due to random chance.
   - Uses test statistics like Z-test, T-test, Chi-square, etc.
   - Widely applied in research, business analytics, and quality control.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
   - **Null hypothesis (H0)**: Assumes no effect or no difference.
   - **Alternative hypothesis (H1)**: Suggests a significant effect or difference.
   - The goal is to test if there is enough evidence to reject H0 in favor of H1.
   - Example: H0: "New drug has no effect," H1: "New drug is effective."
   - Rejection of H0 depends on statistical evidence from sample data.

3. What is the significance level in hypothesis testing, and why is it important?
   - Denoted as alpha (α), typically set at 0.05 (5%).
   - Represents the probability of rejecting a true null hypothesis (Type I error).
   - A lower alpha means stricter evidence is needed to reject H0.
   - Common values: 0.01, 0.05, and 0.10 depending on the field.
   - Helps control the risk of drawing incorrect conclusions.

4. What does a P-value represent in hypothesis testing?
   - The probability of obtaining results as extreme as the observed, assuming H0 is true.
   - A smaller P-value indicates stronger evidence against H0.
   - If P-value < α, we reject H0; otherwise, we fail to reject H0.
   - P-value does not measure effect size or practical significance.
   - Used in all types of hypothesis tests (Z-test, T-test, Chi-square, etc.).

5. How do you interpret the P-value in hypothesis testing?
   - If **P-value < α**: Reject H0, meaning there is significant evidence for H1.
   - If **P-value ≥ α**: Fail to reject H0, meaning there is insufficient evidence for H1.
   - A very small P-value suggests strong evidence against H0.
   - P-value interpretation should consider sample size and context.
   - Misinterpretation can lead to incorrect statistical conclusions.

6. What are Type I and Type II errors in hypothesis testing?
   - **Type I error (False Positive)**: Rejecting H0 when it is actually true (probability = α).
   - **Type II error (False Negative)**: Failing to reject H0 when it is actually false (probability = β).
   - Reducing Type I error increases the risk of Type II error, and vice versa.
   - Example: A test incorrectly concluding that a drug is effective (Type I) or missing an effective drug (Type II).
   - Balance is maintained by selecting an appropriate significance level and sample size.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
   - **One-tailed test**: Tests for an effect in one direction (e.g., H1: "X is greater than Y").
   - **Two-tailed test**: Tests for an effect in both directions (e.g., H1: "X is not equal to Y").
   - A one-tailed test has more statistical power but assumes a directional effect.
   - A two-tailed test is more conservative and widely used in research.
   - Choice depends on prior knowledge and research objectives.

8. What is the Z-test, and when is it used in hypothesis testing?
   - A statistical test used when population variance is known and sample size is large (n > 30).
   - Compares sample mean to population mean.
   - Formula: **Z = (X̄ - μ) / (σ/√n)**.
   - Used for testing proportions and means.
   - Common in quality control, finance, and economics.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
   - **Formula**: Z = (X - μ) / σ.
   - Measures how many standard deviations a value is from the mean.
   - A high absolute Z-score suggests a value significantly differs from the mean.
   - Helps in normalizing data across different distributions.
   - Used in probability calculations and hypothesis testing.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
    - A probability distribution used when the sample size is small (n < 30) and population variance is unknown.
    - More spread out and has thicker tails than the normal distribution.
    - As sample size increases, the T-distribution approaches the normal distribution.
    - Used in T-tests for mean comparisons.
    - Important in small-sample statistical inference.

11. What is the difference between a Z-test and a T-test?
    - **Z-test**: Used when population variance is known and n > 30.
    - **T-test**: Used when population variance is unknown and n < 30.
    - Both compare sample mean to population mean.
    - T-test uses the T-distribution, while Z-test uses the standard normal distribution.
    - T-test is more flexible but requires a smaller sample size assumption.

12. What is the T-test, and how is it used in hypothesis testing?
    - Compares means of two groups when population variance is unknown.
    - Types: One-sample, two-sample, and paired T-tests.
    - Formula: **t = (X̄ - μ) / (s/√n)**.
    - Used in medical trials, A/B testing, and business analytics.
    - Helps determine if mean differences are statistically significant.

13. What is the relationship between Z-test and T-test in hypothesis testing?
    - Both test population mean assumptions.
    - Z-test applies to large samples with known variance, while T-test applies to small samples.
    - Both use critical values to determine statistical significance.
    - In large samples, T-test approximates Z-test results.
    - Both are foundational in inferential statistics.

14. What is a confidence interval, and how is it used to interpret statistical results?
    - A range of values likely to contain the population parameter.
    - Calculated as **CI = X̄ ± Z(σ/√n)** (for known variance) or **CI = X̄ ± t(s/√n)** (for unknown variance).
    - Indicates estimation precision and reliability.
    - A 95% CI means we are 95% confident the true value lies within the range.
    - Used in research, surveys, and forecasting.

15. What is the margin of error, and how does it affect the confidence interval?
    - The maximum expected difference between sample estimate and true population value.
    - Larger sample size reduces the margin of error.
    - Affects the width of the confidence interval.
    - Influenced by confidence level and sample variability.
    - Used in polling and statistical reporting.

16. How is Bayes' Theorem used in statistics, and what is its significance?
    - Describes how to update probabilities with new evidence.
    - Formula: **P(A|B) = (P(B|A) * P(A)) / P(B)**.
    - Used in spam detection, medical diagnostics, and AI.
    - Helps in decision-making under uncertainty.
    - Forms the basis for Bayesian statistics.

17. What is the Chi-square distribution, and when is it used?
    - Used in categorical data analysis.
    - Applies to independence tests and goodness-of-fit tests.
    - Has skewed distribution depending on degrees of freedom.
    - Used in survey analysis and genetics.
    - Compares observed vs. expected frequencies.
18. What is the Chi-square goodness of fit test, and how is it applied?
   - The Chi-square goodness of fit test checks if a sample distribution matches an expected distribution.
   - It is used to determine if observed frequencies differ significantly from expected frequencies.
   - The formula for the test statistic is:
     \[ \chi^2 = \sum \frac{(O - E)^2}{E} \]
     where O is the observed frequency and E is the expected frequency.
   - A higher Chi-square value suggests a greater difference between observed and expected values.
   - Common applications include checking if dice are fair or if customer preferences match expectations.

19. What is the F-distribution, and when is it used in hypothesis testing?
   - The F-distribution is a probability distribution that arises in the analysis of variance (ANOVA) and regression analysis.
   - It is used to compare variances between two or more groups.
   - The F-statistic is given by:
     \[ F = \frac{S_1^2}{S_2^2} \]
     where \( S_1^2 \) and \( S_2^2 \) are sample variances.
   - It is always right-skewed and depends on degrees of freedom.
   - Used in testing whether two populations have the same variance or in ANOVA for testing multiple means.

20. What is an ANOVA test, and what are its assumptions?
   - ANOVA (Analysis of Variance) determines if there are significant differences between the means of multiple groups.
   - It helps test hypotheses about population means without performing multiple t-tests.
   - Assumptions of ANOVA:
     - **Independence**: Observations must be independent of each other.
     - **Normality**: The data in each group should follow a normal distribution.
     - **Homogeneity of variance**: Variances should be equal across groups (checked using Levene’s test).
   - Used in fields like medicine, business analytics, and social sciences to compare group effects.

21. What are the different types of ANOVA tests?
   - **One-way ANOVA**: Compares means of three or more independent groups based on one factor.
   - **Two-way ANOVA**: Examines the effect of two independent factors on a dependent variable.
   - **Repeated measures ANOVA**: Used when the same subjects are tested under different conditions.
   - **MANOVA (Multivariate ANOVA)**: Used when there are multiple dependent variables.
   - Each type is used depending on the study design and number of independent variables.

22. What is the F-test, and how does it relate to hypothesis testing?
   - The F-test is a statistical test used to compare two population variances or to test overall significance in regression and ANOVA.
   - It is based on the F-distribution and compares the ratio of variances.
   - The null hypothesis typically states that variances are equal.
   - If the computed F-value is larger than the critical value from the F-table, the null hypothesis is rejected.
   - Used in comparing model fits in regression and determining differences in group variances.

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test(sample, pop_mean, pop_std):
    sample_mean = np.mean(sample)
    n = len(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    return z_score, p_value

sample_data = np.random.normal(50, 10, 30)
z_stat, p_val = z_test(sample_data, 50, 10)
print(f"Z-Statistic: {z_stat}, P-value: {p_val}")
```

```python
# 2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.
sample = np.random.normal(60, 5, 50)
t_stat, p_value = stats.ttest_1samp(sample, 60)
print(f"T-Statistic: {t_stat}, P-value: {p_value}")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
x = np.linspace(-4, 4, 1000)
pdf = stats.norm.pdf(x)
plt.plot(x, pdf, label="Standard Normal Distribution")
plt.axvline(-1.96, color='red', linestyle='dashed', label="Critical Value -1.96")
plt.axvline(1.96, color='red', linestyle='dashed', label="Critical Value 1.96")
plt.fill_between(x, pdf, where=(x < -1.96) | (x > 1.96), color='red', alpha=0.3)
plt.title("Two-Tailed Z-Test: Decision Regions")
plt.legend()
plt.show()
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
def type1_type2_errors():
    x = np.linspace(-4, 4, 1000)
    pdf = stats.norm.pdf(x)
    plt.plot(x, pdf, label="Standard Normal")

    plt.fill_between(x, pdf, where=(x < -1.96) | (x > 1.96), color='red', alpha=0.3, label="Type I Error")
    plt.fill_between(x, pdf, where=(x > -1.96) & (x < 1.96), color='blue', alpha=0.3, label="Type II Error")

    plt.legend()
    plt.title("Type 1 & Type 2 Errors in Hypothesis Testing")
    plt.show()

type1_type2_errors()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
def confidence_interval(sample, confidence=0.95):
    mean, std_err = np.mean(sample), stats.sem(sample)
    margin = stats.t.ppf((1 + confidence) / 2, len(sample)-1) * std_err
    return mean - margin, mean + margin

sample_data = np.random.normal(100, 15, 40)
ci = confidence_interval(sample_data)
print(f"95% Confidence Interval: {ci}")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
def margin_of_error(sample, confidence=0.95):
    std_err = stats.sem(sample)
    return stats.t.ppf((1 + confidence) / 2, len(sample)-1) * std_err

sample_data = np.random.normal(80, 10, 50)
moe = margin_of_error(sample_data)
print(f"Margin of Error: {moe}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior, likelihood, evidence):
    return (likelihood * prior) / evidence

prior = 0.01  # Probability of having the disease
likelihood = 0.9  # Test accuracy if you have the disease
evidence = (likelihood * prior) + ((1 - prior) * 0.1)  # Total probability

posterior = bayes_theorem(prior, likelihood, evidence)
print(f"Posterior Probability: {posterior}")
```

```python
# 16. Implement an F-test using Python to compare the variances of two random samples.
sample1 = np.random.normal(100, 10, 30)
sample2 = np.random.normal(100, 15, 30)

f_stat = np.var(sample1, ddof=1) / np.var(sample2, ddof=1)
p_value = 1 - stats.f.cdf(f_stat, len(sample1) - 1, len(sample2) - 1)

print(f"F-Test for Variance: F-Statistic = {f_stat}, P-value = {p_value}")
```

```python
# 25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
success_a, n_a = 40, 100
success_b, n_b = 30, 100
p_hat = (success_a + success_b) / (n_a + n_b)
se = np.sqrt(p_hat * (1 - p_hat) * (1/n_a + 1/n_b))
z_score = (success_a/n_a - success_b/n_b) / se
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
print(f"Z-test for Proportions: {z_score}, P-value: {p_value}")
```
